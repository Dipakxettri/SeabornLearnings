A bar plot in Seaborn is atype of plot that shows categorical data using rectangles (bars), where the height of each bar represents a value. Itâ€™s commonly used to compare quantities across categories.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
sns.barplot(data = df, x = "species", y = "body_mass_g", hue = "sex")
```

```python
sns.barplot(data = df, x = "species", y = "body_mass_g", hue = "sex", estimator=np.sum)
```
