# An introduction to seaborn
Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.
Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
df['species'].value_counts()
df['island'].value_counts()
```

```python
sns.set_style("whitegrid") # give styles to the plot
sns.scatterplot(data = df, x = 'flipper_length_mm', y = "body_mass_g", hue = df['island'] )
sns.despine() # remove spines
```

## context

```python
sns.set_style("whitegrid") # give styles to the plot
sns.set_context("poster")
sns.set_context("paper")
sns.set_context("notebook")
sns.scatterplot(data = df, x = 'flipper_length_mm', y = "body_mass_g", hue = df['island'] )
sns.despine() # remove spines
```

## Palette

```python
sns.set_style("whitegrid") # give styles to the plot
sns.set_context("poster")
sns.set_context("paper")
sns.set_context("notebook")
sns.scatterplot(data = df, x = 'flipper_length_mm', y = "body_mass_g", hue = df['island'], palette = "Set3" )
sns.despine() # remove spines
```

```python
sns.scatterplot(data = df, x = "species", y = "body_mass_g", hue = "island", style = "sex", alpha = 0.5)
```

```python
sns.stripplot(data = df, x = "species", y = "body_mass_g", hue = "island", dodge=True,jitter = True)
```

```python
# randomly spread for better readibility
# Swarm plot
sns.set_context("talk")
sns.swarmplot(data = df, x = "species", y = "body_mass_g", hue = "island")
```

```python
# Visualize only one feature

# histogram: 
sns.set_context("notebook")
sns.histplot(data = df, x = "body_mass_g")
```

```python
sns.set_context("notebook")
sns.histplot(data = df, x = "body_mass_g", hue="sex", multiple="stack")
```

```python
# Regression plot :
sns.regplot(data = df, x = "body_mass_g", y = "flipper_length_mm")
```

```python
# line plot:
sns.lineplot(data = df, x = "body_mass_g", y = "flipper_length_mm", hue = "sex")
```

```python
# jointplot :
sns.jointplot(data = df, x = "body_mass_g", y = "flipper_length_mm")
```

```python
sns.jointplot(data = df, x = "body_mass_g", y = "flipper_length_mm", hue = "sex")
```

```python
sns.jointplot(data = df, x = "body_mass_g", y = "flipper_length_mm", hue = "sex", kind = "kde")
```

```python
# bar plot :
sns.barplot(data = df, x = "species", y = "body_mass_g", hue ="sex", palette=["pink","skyblue"])
```

```python
# Count plot

sns.countplot(data = df, x  = "species", hue = "island")
```

```python
# Box plot

sns.boxplot(data = df, x = "species", y ="body_mass_g", hue = "island")
```

```python
# violin plot

sns.violinplot(data = df, x = "species", y = "body_mass_g",hue = "sex")
```

```python
sns.violinplot(data = df, x = "species", y = "body_mass_g",hue = "sex", split = True)
```

```python
sns.violinplot(data = df, x = "species", y = "body_mass_g",hue = "sex",split = True, inner = "box")
```

```python
# violint plot + swarmplot
sns.violinplot(data = df, x = "species", y = "body_mass_g", palette="pastel")
sns.swarmplot(data = df, x = "species", y = "body_mass_g", hue = "island")
```

```python
# KDE plot :

sns.kdeplot(data = df, x = "body_mass_g", hue = "species", palette = "pastel",fill = "True")
```

```python
# Heat Map :

columns = ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]

sns.heatmap(data = df[columns].corr(), annot = True, cmap = "coolwarm")
```

```python
sns.heatmap(data = df[columns].corr(), annot = True, cmap = "Blues")
```
