A regression plot in Seaborn is a plot that shows the relationship between two numeric variables and automatically fits a regression line to the data.
Think of it as a scatter plot with a best-fit line.

A linear regression line is a straight line that models the relationship between two numeric variables by minimizing the distance between the line and all the data points.
In simple words: it’s the “best-fit line” that predicts y from x.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
sns.regplot(data = df, x = "bill_depth_mm", y = "body_mass_g")
```
