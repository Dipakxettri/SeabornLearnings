A strip plot in Seaborn is a type of plot that shows all individual data points for a categorical variable. 
Think of it as a 1-dimensional scatter plot along a categorical axis.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# first loading data :
sns.get_dataset_names()

df = sns.load_dataset("penguins")
df
```

## Plotting a strip plot

```python
sns.stripplot(data = df, x = "species", y = "body_mass_g")
```
