```python
# A histogram in Seaborn is a plot that shows the distribution of a numeric variable by dividing the data into bins and counting how many observations fall into each bin.
# Think of it as a way to see how your data is spread out.
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# first loading data :
sns.get_dataset_names()

df = sns.load_dataset("penguins")
df
```

```python
sns.histplot(data=df, x = "body_mass_g", hue = "sex")
```

## multiple

What multiple Does

When you use hue to separate data into categories, multiple decides how the bars for each category are arranged.

⭐ Options for multiple
Option	Description	Example
"layer"	Default. All bars overlap but with transparency.	Bars stacked on top of each other visually.
"stack"	Bars are stacked on top of each other vertically.	Total height = sum of counts across groups.
"dodge"	Bars are side by side for each category.	Each group gets its own bar next to others.
"fill"	Bars are stacked and normalized to 100%.	Shows proportion of each group instead of count.

⭐ Example with multiple
sns.histplot(
    data=df,
    x="body_mass_g",
    hue="species",
    bins=20,
    multiple="stack",   # try "dodge", "fill", "layer"
    palette="Set2"
)

How it looks:
layer → transparent overlapping bars
stack → stacked bars, sum = total count
dodge → side-by-side bars for each hue category
fill → stacked bars scaled to 100% (proportions)

⭐ Tip
Use "stack" to see total counts with breakdowns.
Use "dodge" to compare groups side by side.
Use "fill" to compare relative proportions.
Use "layer" for a visual overlay if colors are semi-transparent.

```python
sns.histplot(data = df, x = "body_mass_g", hue="sex", multiple="stack")
```
