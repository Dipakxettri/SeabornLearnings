A swarm plot in Seaborn is very similar to a strip plot,
but it automatically adjusts the positions of points to avoid overlap.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# first loading data :
sns.get_dataset_names()

df = sns.load_dataset("penguins")
df
```

```python
sns.swarmplot(data = df, x = "species", y = "body_mass_g", hue = "sex")
```
