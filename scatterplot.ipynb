```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# first loading data :
sns.get_dataset_names()

df = sns.load_dataset("penguins")
df
```

```python
# scatter plot:

sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g")
```

## hue

It tells Seaborn to color the points (or bars/lines/etc.) based on the unique categories inside a column.

```python
# usng hue 
sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g", hue = "sex")
```

## Color & Palettes

```python
# Single color:

sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g", color="red")
```

```python
# Categorical palette:

# palette="Set2"
# palette="tab10"
# palette="Pastel1"

sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g", color="red",hue = "sex", palette="Set2")
```

```python
# Sequential palette:

# palette="viridis"
# palette="magma"
# palette="plasma"

sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g", color="red",hue = "sex", palette="viridis")
```

```python
# Diverging palette:

# palette="coolwarm"
# palette="RdBu"

sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g", color="red",hue = "sex", palette="coolwarm")
```

```python
# Custom palette:

# palette=["#ff0000", "#00ff00", "#0000ff"]

sns.scatterplot(data = df, x = "flipper_length_mm", y = "body_mass_g", color="red",hue = "sex", palette=["#ff0000","#00ff00"])
```
