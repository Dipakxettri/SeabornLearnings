A count plot in Seaborn is a plot that shows the count of observations for each categorical value. Itâ€™s essentially a bar plot of frequencies, where the height of each bar represents how many times that category appears in the dataset.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
sns.countplot(data = df, x = "island", hue = "species")
```
