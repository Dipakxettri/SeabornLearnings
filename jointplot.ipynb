A jointplot in Seaborn is a powerful plot that combines a scatter plot (or other kind of plot) with univariate distributions for two variables.

It lets you see the relationship between two numeric variables along with their marginal distributions at the top and right sides.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
sns.jointplot(data = df, x = "body_mass_g", y = "flipper_length_mm")
```

## Hue 

Colors data points (or distributions) according to the sex column.can also change a second type of diagram

```python
sns.jointplot(data = df, x = "body_mass_g", y = "flipper_length_mm", hue = "sex")
```

## Kind 

The kind parameter tells Seaborn what kind of main plot to draw.

```python
sns.jointplot(data = df, x = "body_mass_g", y = "flipper_length_mm", hue = "sex", kind = "kde")
```
